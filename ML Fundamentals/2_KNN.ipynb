```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

train_data = pd.read_csv("celebs_train.csv")
X_train = train_data[['Height', 'Weight']]
y_train = train_data.Gender

test_data = pd.read_csv('celebs_test.csv')
X_test = test_data[['Height', 'Weight']]
y_test = test_data.Gender

plt.scatter(train_data.Height, train_data.Weight, c = train_data.Gender)
plt.title('Weight vs. Height (Training Data)')
plt.ylabel('Weight')
plt.xlabel('Height')
plt.show()

print('train_data: \n', train_data)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def knn(k):
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(X_train, y_train)

  y_pred = classifier.predict(X_test)
  test_data['pred'] = y_pred

  plt.scatter(train_data.Height, train_data.Weight, c = train_data.Gender)
  plt.scatter(test_data[test_data.pred==0].Height, test_data[test_data.pred==0].Weight, c = 'r', label ='test_pred_0')
  plt.scatter(test_data[test_data.pred==1].Height, test_data[test_data.pred==1].Weight, c = 'b', label ='test_pred_1')
  plt.title('Weight vs. Height')
  plt.ylabel('Weight')
  plt.xlabel('Height')
  plt.legend()
  plt.show()

  print('test_data: \n', test_data)
```

```python
knn(k=3)
```

```python
knn(k=5)
```
